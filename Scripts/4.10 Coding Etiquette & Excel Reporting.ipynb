# Step 3

```python
# Creating an empty list to store the regions

region=[]
```

```python
# Checking the format of states

customer['state'].value_counts()
```

```python
# Populating the regions

for x in customer['state']:
        if x in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']:
            region.append("northeast")
        elif x in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']:
            region.append("midwest")
        elif x in ['Delaware','Maryland','District of Columbia','Virginia','West Virginia','North Carolina','South Carolina','Georgia','Florida','Kentucky','Tennessee','Mississippi','Alabama','Oklahoma','Texas','Arkansas','Louisiana']:
            region.append("south")
        else:
            region.append("west")
```

```python
region
```

```python
# Adding region column to the customer data frame

customer['region']= region
```

```python
customer['region'].value_counts()
```

```python
# Creating a crosstab between region and spending flag

region_tab = pd.crosstab(customer['region'],customer['spending_flag'], dropna=False)
```

# Step 5

```python
# Creating the flag for age

age_flag=[]

for i in customer['age'].to_list():
    if i<=30:
        age_flag.append("Young")
    elif i>30 and i<65:
        age_flag.append("Middle age")
    elif i>=65:
        age_flag.append("Senior")
    else:
        age_flag.append("invalid")
```

```python
customer['age_flag']=age_flag
```

```python
customer['age_flag'].value_counts()
```

```python
# Creating the flag for income

income_flag=[]

for i in customer['income'].to_list():
    if i<=75000:
        income_flag.append("Low income")
    elif i>75000 and i<150000:
        income_flag.append("Middle income")
    elif i>=150000:
        income_flag.append("High income")
    else:
        income_flag.append("invalid")
```

```python
customer['income_flag']=income_flag
```

```python
customer['income_flag'].value_counts()
```

```python
# Crosstab the eating habits of customers

dep_cross = pd.crosstab(customer['department_id'],customer['user_id'],dropna=False)
```

```python
dep_cross
```

```python
# Segregating the customers based on their eating habits that is vegan and non-vegan

vegan=dict()
for user in dep_cross:
    working=dep_cross[user].tolist()
    if working[11]==0 and working[15]==0:
        vegan[user]='Vegan'
    else:
        vegan[user]='Non vegan'
```

```python
vegan_flag=[]
for user in customer['user_id']:
    vegan_flag.append(vegan[user])
```

```python
# Adding the flag of vegan to the customer data

customer['vegan_flag']=vegan_flag
```

```python
customer['vegan_flag'].value_counts()
```
