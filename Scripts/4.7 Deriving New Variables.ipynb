# Step 4

```python
# Countiung the values 

count=df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# converting counts to dictionary 

count_dict=count.to_dict()
```

```python
# Finding the quantile range

count_quant=list(count.quantile([0.75,0.5,0.25]))
```

```python
count_quant
```

```python
result=[]
for value in df_ords_prods_merge['order_hour_of_day']:
    for key,val in count_dict.items():
        if value==key:
            if val<count_quant[2]:
                result.append("Fewest orders")
            elif val>= count_quant[2] and val< count_quant[1]:
                result.append("Average orders")
            elif val>=count_quant[1]:
                result.append("Most orders")
```

```python
# Printing the number of orders 

def labelling(i):
    most_ord=i.count("Most orders")
    avg_ords=i.count("Average orders")
    least=i.count("Most orders")
    print("most orders ",most_ord)
    print("Fewest orders ",avg_ords)
    print("Least number of orders ",least)
    print("Total number of orders ",most_ord+avg_ords+least)
```

```python
labelling(result)
```

```python
# Creating another colomns (busiest_period_of_day)

df_ords_prods_merge['busiest_period_of_day']=result
```
